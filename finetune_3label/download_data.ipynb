```python
import requests

urls = {
    "train.tsv": "https://raw.githubusercontent.com/indobenchmark/indonlu/master/dataset/smsa_doc-sentiment-prosa/train_preprocess.tsv",
    "valid.tsv": "https://raw.githubusercontent.com/indobenchmark/indonlu/master/dataset/smsa_doc-sentiment-prosa/valid_preprocess.tsv",
    "test.tsv": "https://raw.githubusercontent.com/indobenchmark/indonlu/master/dataset/smsa_doc-sentiment-prosa/test_preprocess.tsv"
}

# download
for filename, url in urls.items():
    r = requests.get(url)
    with open(filename, "wb") as f:
        f.write(r.content)
    print(f"{filename} downloaded!")
```
